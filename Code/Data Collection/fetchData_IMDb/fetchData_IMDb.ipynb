```python
def parse_without_selenium(url = url, headers = headers):
    r = requests.get(url, headers = headers)
    soup = BeautifulSoup(r.text, "html.parser")
    # print(soup.prettify())
    return soup
```

```python
def collect_data(soup = soup, url = url):

    # print(soup.prettify())
    # print(soup.find('div', {"data-testid" : "hero-rating-bar__aggregate-rating__score"}))
    #IMDb rating
    # imdb_rating = soup.find('div', {"data-testid" : "hero-rating-bar__aggregate-rating__score"}).text
    # imdb_rating = float(imdb_rating.replace('/10',''))

    
    try:
        imdb_rating = soup.find('div', {"data-testid" : "hero-rating-bar__aggregate-rating__score"}).text
        imdb_rating = float(imdb_rating.replace('/10',''))
    except:
        imdb_rating = None
        
    try:
        budget = soup.find('li', {"data-testid" : "title-boxoffice-budget"}).text
        budget = int(''.join(filter(str.isdigit, budget)))
    except AttributeError:
        budget = None
    
    try:
        gross = soup.find('li', {"data-testid" : "title-boxoffice-cumulativeworldwidegross"}).text
        gross = int(''.join(filter(str.isdigit, gross)))
    except AttributeError:
        gross = None
    
    try:
        aspect_ratio = soup.find('li', {"data-testid" : "title-techspec_aspectratio"}).text
    except AttributeError:
        aspect_ratio = None

    data_list = [imdb_rating, budget, gross, aspect_ratio, url]
    # print(data_list)
    
    return data_list

    

```

```python
def list_iteration(url_list = url_list, headers = headers):
    final_df = pd.DataFrame()
    for index, url in enumerate(url_list):
        
        random_number = str(random.randint(100, 999))
        random_number2 = str(random.randint(100,999))
        headers = {
        'User-Agent': f'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/{random_number2}.36 (KHTML, like Gecko) Chrome/{random_number}.0.0.0 Safari/{random_number2}.36'
        }
        print(index)
        print(url)

        soup = parse_without_selenium(url, headers)
        data_list = collect_data(soup, url)
        df_row = pd.DataFrame(data_list)
        
        final_df = pd.concat([final_df, df_row], axis=1)
        if index%100 == 0:
            final_df.transpose().to_csv(r'C:\Users\abhyu\Desktop\Dissertation\Code\fetchData_IMDb\backup_data.csv', index= False)
        #Testing
        # if index == 3:
        #     break
    return final_df.transpose()
```
