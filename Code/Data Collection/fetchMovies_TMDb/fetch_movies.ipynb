```python
#define values
vote_count_gte = 100
start_year = 2012
```

```python
#iterating through pages
def iterate_data_for_year(total_pages, url):

    for count in range(1, total_pages):
        page = url + str(count)
        page_data = requests.get(page).json()
        # print(page_data['results'])
        for film in page_data['results']:
            print(film['title'] + ", "+film['release_date'])

```
