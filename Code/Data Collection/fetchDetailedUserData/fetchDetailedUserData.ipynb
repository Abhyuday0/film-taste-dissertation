```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36'
}

data = pd.read_csv(r"C:\Users\abhyu\Desktop\Dissertation\DATA\popular_user_data.csv")

user_list = data["username"].to_list()
url = 0
```

```python
def parse(url = url, headers = headers):
    params = {'sort': 'date', 'view': 'list', 'hide': 'shorts,tv'}
    r = requests.get(url, headers = headers, params=params)
    soup = BeautifulSoup(r.text, "html.parser")
    # print(soup.prettify())
    return soup
```

```python
def get_user_data(username, headers):
    # Initialize empty list to store films
    films = []

    # Loop through each page of the user's films and extract information
    page_num = 1
    # films = []
    
    while True:
        soup = parse(f"https://letterboxd.com/{username}/films/diary/page/{page_num}/", headers)

        
        films_on_page = soup.find_all('tr', class_="diary-entry-row")
        # end the loop
        if len(films_on_page) == 0:
            break
        
        
        for film in films_on_page:
            film_details_element = film.find('td', {'class' : 'td-film-details'})
            film_details = film_details_element.find('div')
            film_id = film_details['data-film-id']
            film_url = 'https://letterboxd.com' + film_details['data-film-slug']
            
            
            try:
                film_rating_element = film.find('div', class_="hide-for-owner")
                film_rating = film_rating_element.find('span')['class'][1]
                film_rating = int(film_rating.replace('rated-', ''))
            except:
                film_rating = None

            try:
                film_review = film.find('a', class_="edit-review-button has-icon icon-16 icon-edit")['data-review-text']
            except:
                film_review = None

            if film.find('a', {"data-rewatch":'true'}):
                film_rewatched = 'yes'
            else:
                film_rewatched = 'no'

            film_user_data = {
                            'id' :film_id,
                            'rating': film_rating,
                            'url' : film_url,
                            'review': film_review,
                            'rewatched' : film_rewatched
                        }
            
            # print(film_user_data)
            films.append(film_user_data)
        

        #Testing
        if page_num == 10:
            break
        
        # Move on to the next page
        page_num += 1
    # print(films)
    return films
```

```python
# Make a GET request to the user's films 

def get_data(user_list, headers):

    # test
    # df = get_user_data('Wisenheimer', headers)

    df = pd.DataFrame()

    for index, user in enumerate(user_list):
        user_diary_data = get_user_data(user, headers)
        user_data = [user, user_diary_data, len(user_diary_data)]
        print(index)

        df_row = pd.DataFrame(user_data)
        
        df = pd.concat([df, df_row], axis=1)
        #trial
        # print(df_row)

        # if index == 2:
        #     break
    df = df.transpose()
    df.columns = ['username', 'movies watched', 'number of movies']
    return df
    
        
```

```python
df = get_data(user_list, headers)
```
