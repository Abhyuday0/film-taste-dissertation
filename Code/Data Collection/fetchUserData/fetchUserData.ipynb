```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36'
}
```

```python
#define variables

url = "https://letterboxd.com/members/popular/this/week/page/"
start_page = 5
page_count = 25
soup = 0
link = 0
```

```python
def parse_page(url = url, headers = headers):
    r = requests.get(url, headers = headers)
    soup = BeautifulSoup(r.text, "html.parser")
    # print(soup.prettify)
    return soup
```

```python
def collect_data(soup = soup):

    user_content = soup.find('table', class_="person-table")
    list_of_users = user_content.find_all('a', class_="avatar -a40")

    df_page = pd.DataFrame()
    
    

    #One movie at a time
    for index, attributes in enumerate(list_of_users):
        # print(attributes.get('data-film-name')+', '+ attributes.get('data-film-release-year'))
        
        link = attributes.get('href')
        user_link = 'https://letterboxd.com' + link

        username = link.strip('/')
        print(username)
        
        row = [username, user_link]
        
        # print(row)
        df_row = pd.DataFrame(row)

        df_page = pd.concat([df_page, df_row], axis=1)


    
    
    return df_page
```

```python
def get_data(url=url, page_count=page_count, start_page = start_page, headers = headers):

    df = pd.DataFrame()
    
    for index, count in enumerate(range(start_page, page_count)):
        # print(url)
        # print(str(count))
        print(count)
        
        page_url = url + str(count)
        print(page_url)
        soup = parse_page(page_url, headers)
        df_page = collect_data(soup)
        df = pd.concat([df, df_page.transpose()])
       
        

        # testing
        # if index == 3:
        #     break
        
    return df
```

```python
df = get_data(url, page_count, start_page, headers)
```
